```python
import os
from ultralytics import YOLO

import random as rnd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model = YOLO(model="yolov8n.yaml")
```

```python
epochs = 100
batch = 4

metrics = model.train(
    data = os.path.join("datasets", "coco128", "data.yaml"),
    
    val = True,

    cache = True,
    device = 0,

    epochs = epochs,
    batch = batch,
    optimizer = "Adam",
    lr0 = 0.001,
    lrf = 0.01,
    momentum = 0.9,

    imgsz = 640,
    translate = 0.1,
    scale = 0.5,
    mosaic = 1.0,
    mixup = 0.0,

    save = True,
    save_period = 1,

    seed = 0,
    exist_ok = True,
    project = "train",
    name = "yolov8n_coco128", 
)
```

```python
model = YOLO(os.path.join("train", "yolov8n_coco128", "weights", "last.pt"))

results = model.train(resume = True)
```

```python
model = YOLO(os.path.join("train", "yolov8n_coco128", "weights", "best.pt"))
```

```python
batch = 4
conf = 0.001
iou = 0.6
max_det = 300

metrics = model.val(
    data = os.path.join("datasets", "coco128", "data.yaml"),

    split = "test",

    batch = batch,
    conf = conf,
    iou = iou,
    max_det = max_det,

    exist_ok = True,
    project = "eval",
    name = "yolov8n_coco128",
)
```

```python
print('mAP50-95:', metrics.box.map)
print('mAP50:', metrics.box.map50)
print('mAP75:', metrics.box.map75)
print('mAP50-95 per class:', metrics.box.maps)
```

```python
img = cv2.imread(os.path.join("eval", "yolov8n_coco128", "confusion_matrix.png"))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.title("Original")
plt.show()
```

```python
model = YOLO(os.path.join("train", "yolov8n_coco128", "weights", "best.pt"))
```

```python
results = model.predict(
    source = os.path.join("datasets", "coco128", "test", "images"),
    conf = 0.1,
    iou = 0.5,
    max_det = 300,
    save_txt = True,
    save = True,
    verbose = False,
    exist_ok = True,
    project = "predict",
    name = "yolov8n_coco128",
)
```

```python
index = rnd.randint(0, len(results))
img = cv2.imread(results[index].path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
boxes = results[index].boxes

for box in boxes:
    x1, y1, x2, y2 = box.xyxy[0]
    label = results[0].names[int(box.cls[0])]
    
    cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 255), 2)
    cv2.putText(img, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 2)

plt.imshow(img)
plt.axis('off')
plt.title("Image w/ Detections")
plt.show()
```

```python
predict_dir = os.listdir(os.path.join('predict', 'yolov8n_coco128'))

img_det = cv2.imread(os.path.join('predict', 'yolov8n_coco128', predict_dir[index]))
img_det = cv2.cvtColor(img_det, cv2.COLOR_BGR2RGB)

plt.imshow(img_det)
plt.axis('off')
plt.title("Image w/ Detections")
plt.show()
```
